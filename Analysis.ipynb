```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.animation as animation
```

```python
data = pd.read_csv("Simul.txt", sep=" ", header=None)
```

```python
time = 1.0
dt = 0.0001
steps = int(time/dt) 
particle_num = 512

x_array = np.zeros((steps, particle_num))
y_array = np.zeros((steps, particle_num))
z_array = np.zeros((steps, particle_num))
```

```python
print(range(0,5))
```

```python
c = 0
for i in range(0,steps):
    for j in range(0,particle_num):
        x_array[i][j] = data[0][c]
        y_array[i][j] = data[1][c]
        z_array[i][j] = data[2][c]
        c = c+1
```

```python
start = 0
stop = 9999
part_num = 201
posx = np.zeros(stop-start)
posy = np.zeros(stop-start)

for i in range(0, stop-start):
    posx[i] = x_array[i][part_num]
    posy[i] = y_array[i][part_num]

ax = plt.axes()
ax.scatter(0.5, 0.5, color="red", s = 0.9)
ax.plot(posx, posy)
ax.set_ylim(0.0,1.0)
ax.set_xlim(0.0,1.0)
plt.gca().set_aspect(1.0)
plt.show()
```

```python
frame = 9999
a = 0.0

fig = plt.figure()
ax = plt.axes(projection = "3d")

ax.scatter(x_array[frame], y_array[frame], z_array[frame], s=1.5)
ax.view_init(elev = 90)
ax.set_xlim(a,a+1.0)
ax.set_ylim(a,a+1.0)
ax.set_zlim(a,a+1.0)
plt.show()
```

```python
a = 0.0
num_frames = 999
num_fps = 40

fig = plt.figure()  
ax = plt.axes(xlim=(a, a+1.0), ylim=(a, a+1.0))
plt.gca().set_aspect('equal')
   
# initializing a line variable 
scat = ax.scatter([], [], s = 0.8)  
   
def animate(i): 
    i = i*10
    x = x_array[i]
    y = y_array[i] 
    
    arr = np.zeros((particle_num, 2))
    for i in range(0,particle_num):
        arr[i] = [x[i], y[i]]
    
    scat.set_offsets(arr)
    return scat,
   
anim = animation.FuncAnimation(fig, animate, frames = num_frames, interval = 50, blit = True)
  
anim.save('GravSimulationXY.gif',  
          writer = 'Pillow', fps = num_fps) 
```
